```python
# plotting distplot and boxplot for the non-object variables

var=['SepalLengthCm',	'SepalWidthCm',	'PetalLengthCm',	'PetalWidthCm',	'Species']
def Univariate(var):
  fig,ax= plt.subplots(1,2,figsize=(20,5))
  sns.distplot(var,kde=True,ax=ax[0])
  sns.boxplot(var,ax=ax[1])
```

```python
#plotting bar plot and pie chart for the categorical variable in the dataset

def univariate_categ(var):
  fig,ax  = plt.subplots(1,2,figsize=(20,8))
  cat=sns.countplot(var,ax=ax[0])
  cat.set_xticklabels(list(cat.get_xmajorticklabels()),rotation=45)
  var.value_counts().plot.pie(ax=ax[1],autopct='%.2f %%')
```

```python
cat_vars=df.select_dtypes('object').columns
```

```python
for var in cat_vars:
  cat_data = df[var]
  univariate_categ (cat_data)
```

## Modelling and Evaluation

```python
#Checking the number of clusters with the help of Elbow Method

x= df.iloc[:,[0,1,2,3]].values

from sklearn import datasets
from sklearn.cluster import KMeans
sum_squares=[]

for i in range(1,15):
  kmeans= KMeans(n_clusters=i, init ='k-means++',max_iter=300, n_init=10, random_state=100)
  kmeans.fit(x)
  sum_squares.append(kmeans.inertia_)
```

```python
#plotting the elbow method plot

plt.figure(figsize=(10,8))
plt.plot(range(1,15),sum_squares)
plt.title('Elbow Method')
plt.show()
```

##### An elbow bent is observed between 2 and 4, hence taking the number of clusters to be 3

```python
#taking the cluster number to be 3 and plotting the clsutering diagram

kmeans = KMeans(n_clusters=3, init = 'k-means++', max_iter=300, n_init = 10, random_state=100)
y_kmeans = kmeans.fit_predict(x)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x[y_kmeans == 0,0],x[y_kmeans == 0,1],s = 100, c='yellow', label='Iris-Setosa')
plt.scatter(x[y_kmeans == 1,0], x[y_kmeans == 1,1], s=100, c='red', label ='Iris-versicolour')
plt.scatter(x[y_kmeans==2,0], x[y_kmeans == 2,1],s=100,c='green', label='Iris-virginica')

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='blue', label='Centroids')
plt.legend()
plt.show()
```
